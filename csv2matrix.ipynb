```python
import pandas 
import matplotlib.pyplot as plt

header_list=[] 
for i in range(100,7500,200): 
    header_list.append(i) 
df = pandas.read_csv('C:/Users/orlov/Desktop/matrix_temp.csv', header=None)

f = plt.figure(figsize=(14, 10)) 
plt.matshow(df, fignum=f.number) 
plt.xticks(range(0, df.shape[1]-1, 2), header_list, rotation=90) 
plt.yticks(range(0, df.shape[1], 2), header_list)
plt.grid()
cb = plt.colorbar() 
cb.ax.tick_params(labelsize=14) 
plt.title('Robinson-Foulds compatibility matrix', fontsize=16);
```
